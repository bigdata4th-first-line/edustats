#### 진학 비율, 사교육비, 음주율, 우울감경험률, 흡연율, 비만율을 Indicator 컬럼으로 넣기

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/18_22비만율추가.csv') # 비만율까지 추가되어 있는 데이터프레임
df
```

```python
df2 = pd.DataFrame() # 빈 데이터프레임 생성

columns = ['특성화고진학률','특목고진학률','자율고진학율','특+자 진학률','사교육비','음주율','우울감경험률','흡연율','비만율']
values = ['특성화고','특목고','자율고']
i = 0

for i in df.index:
    for j in range(0, 9):
        k = i*9 + j
        df2.loc[k,'지역'] = df.loc[i,'지역']
        df2.loc[k,'Indicator Name'] = f'{columns[j]}'
        df2.loc[k,'year'] = str(df.loc[i,'year'])
        if j <= 2:
            df2.loc[k,'value'] = df.loc[i,f'{values[j]}']/df.loc[i,'졸업자']
        elif j == 3:
            df2.loc[k,'value'] = (df.loc[i,f'{values[j-2]}'] + df.loc[i,f'{values[j-1]}'])/df.loc[i,'졸업자']
        elif j == 4:
            df2.loc[k,'value'] = df.loc[i,'사교육비']
        elif j == 5:
            df2.loc[k,'value'] = df.loc[i,'음주율']
        elif j == 6:
            df2.loc[k,'value'] = df.loc[i,'우울감경험률']
        elif j == 7:
            df2.loc[k,'value'] = df.loc[i,'흡연율']
        elif j == 8:
            df2.loc[k,'value'] = df.loc[i,'비만율']
    i += 1

df2
```

```python
# year 컬럼 값을 정수로.
df2['year'] = df2['year'].apply(lambda x: int(x)) 
```

```python
# 'Indicator18_22.csv'로 저장.
df2.to_csv('Indicator18_22.csv', index = False) 
```
